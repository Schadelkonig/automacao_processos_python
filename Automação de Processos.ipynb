## Buscar informações de ações

```python
import yfinance
```

```python
ticker = input("Digite o código da ação: ")

dados = yfinance.Ticker(ticker).history("6mo")
```

```python
codigo = "PETR4SA"
fechamento = dados.Close
```

```python
fechamento
```

```python
# Gerar um gráfico para análise

fechamento.plot()
```

## Gerar análises

#### • Últimos 6 meses
#### • Cotação máxima
#### • Cotação mínima
#### • Cotação atual

```python
maxima = round(fechamento.max(), 2)
```

```python
minima = round(fechamento.min(), 2)
```

```python
atual = round(fechamento[-1], 2)
```

```python
print(maxima)
print(minima)
print(atual)
```

## Enviar email automaticamente

#### • Abrir uma nova aba (Teclar Ctrl + t)
#### • Digitar o endereço do Gmail
#### • Teclar o ENTER
#### • Clicar no botão Escrever
#### • Preencher o destinatário (Teclar Tab)
#### • Preencher assunto (Teclar Tab)
#### • Preencher o corpo do email
#### • Clicar no botão Enviar

```python
import pyautogui
import pyperclip
```

```python
# Dar uma pausa entre os passos
pyautogui.PAUSE = 6

# Abrir uma nova aba (Ctrl + t)
pyautogui.hotkey("ctrl", "t")

# Digitar o endereço do Gmail
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

# Clicar no botão escrever
pyautogui.click(x=97, y=194)

# Preencher o destinatário (teclar tab)
pyperclip.copy("wl.williamlenon@gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# Preencher assunto (teclar tab)
pyperclip.copy("Análises diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# Preencher o corpo do email
mensagem = f"""
Prezado Gestor,

Seguem as análises diárias da ação {codigo} referentes aos últimos seis meses:

Cotação máxima: R$ {maxima}
Cotação mínima: R$ {minima}
Cotação atual: R$ {atual}

Qualuqer dúvida, estou à disposição!
"""
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

# Enviar o email
pyautogui.click(x=842, y=697)
```
